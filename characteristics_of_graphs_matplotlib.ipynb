```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = [1,2,3,4,5,6]
y = [1,2,3,4,5,6]
x2 = [5,5,5,5,5,5]
```

```python
plt.plot(x1,y, color = "green", linestyle ='--',linewidth = 3, alpha = 0.8, label = 'x1')
plt.plot(x2,y, color = "blue", linestyle ='-',linewidth = 1, alpha = 1, label = 'x2')
plt.show()
```
